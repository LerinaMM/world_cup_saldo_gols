**1. Bibliotecas**

```python
import pandas as pd
```

```python
results = pd.read_csv('/content/results.csv', parse_dates=['date']) 
```

```python
results
```

**2. Informações do Dataset**

```python
results.info()
```

3. Realizando mudança do formato de data que está como string inserir após o endereço do arquivo essa etapa **parse_dates=['date'])** para mudar para formato de data  

4. Mudança para o formato atual de data dd/mm/yyyy para deixar somente o ano

```python
results_fifa = results[results['tournament'] == 'FIFA World Cup']
```

```python
results_fifa
```

```python
results_fifa['year'] = results_fifa['date'].dt.year
```

5. Vamos escolher as colunas que nos interessam e mudar o nome dessas colunas

```python
home_scores = results_fifa[['home_team', 'home_score', 'year']].rename({'home_team':'team','home_score':'score'}, axis='columns')
away_scores = results_fifa[['away_team', 'away_score', 'year']].rename({'away_team':'team','away_score':'score'}, axis='columns')
```

```python
home_scores
```

```python
away_scores
```

7. Vamos realizar a concatenação das colunas home_scores e away_scores pois nao importam nesta analise se o saldo de gols é dentro ou fora de casa, além disso vamos realizar um reset do index pois a primeira coluna nao é necessária. Uma terceira operação será substituir os valores de NaN por zero utilizando o comando . fillna(0)

```python
scores_fifa = pd.concat([home_scores, away_scores]).reset_index(drop=True).fillna(0)
```

```python
scores_fifa
```

8. Vamos neste momento realizar o acumulado de saldo de gols por ano e utilizando a função unstack()  
9. Mudar o formato da tabela de preferencialmente utilizando linhas para os anos para utilizar colunas para os anos.
10. No final utilizaremos a função fillna(0) para substituir os valores de NaN por zero

```python
scores_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0)
```

11. Nesta etapa estamos transformando a etapa anterior adicionando a função cumsum(axis=1) onde realiza uma soma das colunas para termos os valores acumulados durante a passagem dos anos

```python
scores_fifa_year = scores_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
```

```python
scores_fifa_year
```

12. Vamos realizar a upload do arquivo countries-fifa-flags.csv onde neste arquivo teremos as informações dos Paises e Bandeiras

```python
flags = pd.read_csv('/content/countries-fifa-flags-1.csv')
```

```python
flags
```

13. Vamos realizar a junção dos dois Dataframes: Resultados e Countries-fifa_flags-1 e além disso vamos transformar esses resultados para um csv, adicionando **to_csv('fifa-data-viz.csv', index='False')**

```python
scores_fifa_year.merge(flags, left_on='team', right_on='country', how='left').to_csv('fifa-data-viz.csv', index=False)
```
